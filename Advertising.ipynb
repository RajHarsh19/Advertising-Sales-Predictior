```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Advertising.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
X=df.drop('sales',axis=1)
y=df['sales']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
X_validat, X_holdout_test, y_validat, y_holdout_test = train_test_split(X_test, y_test, test_size=0.5, random_state=101)
```

```python
len(X)
```

```python
len(X_train)
```

```python
len(X_validat)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model=RandomForestRegressor(n_estimators=30,random_state=101)
```

```python
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
validat_predictions=model.predict(X_validat)
```

```python
validat_predictions
```

```python
mean_absolute_error(y_validat,validat_predictions)
```

```python
np.sqrt(mean_squared_error(y_validat,validat_predictions))
```

```python
holdout_predictions = model.predict(X_holdout_test)
```

```python
holdout_predictions
```

```python
mean_absolute_error(y_holdout_test,holdout_predictions)
```

```python
np.sqrt(mean_squared_error(y_holdout_test,holdout_predictions))
```

```python
final_model = RandomForestRegressor(n_estimators=30,random_state=101)
```

```python
final_model.fit(X,y)
```

```python
final_model.predict(X_test)
```

```python
import h5py
import pickle
```

```python
with h5py.File("advertising.h5", "w") as f:
    model_bytes = pickle.dumps(final_model)  # Convert model to bytes
    f.create_dataset("model", data=np.void(model_bytes))  # Store in HDF5 format

print("Model saved as advertising.h5")
```
